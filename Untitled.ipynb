```python
import pandas as pd
```

```python
df = pd.read_csv('Telco.csv')
```

```python
df.head()
```

```python
# Create Customer_Num and Customer_Xter from customerID
# Label encording for PaymentMethod
#Customer_Num ** 2
```

```python
df['PaymentMethod'].unique()
```

```python
Payment_code = {'Electronic check': 1, 'Mailed check': 2, 'Bank transfer (automatic)': 3,'Credit card (automatic)': 4}
```

```python
df['PaymentMethod'] = df['PaymentMethod'].apply(lambda x:Payment_code[x])
```

```python
df.head()
```

```python
df['PaymentMethod'] = df['PaymentMethod'].apply(lambda x:Payment_code.get(x, 'not found'))
```

```python
df.head()
```

```python
df['Customer_Num'] = df['customerID'].apply(lambda x : x.split('-')[0])
```

```python
df['Customer_Xter'] = df['customerID'].apply(lambda x : x.split('-')[1])
```

```python
df.head()
```

```python
df['square'] = df['Customer_Num'].apply(lambda x : int(x) ** 2)
```
